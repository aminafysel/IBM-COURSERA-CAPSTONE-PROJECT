```
import numpy as np

import matplotlib.pyplot as plt

from google.colab import files
uploaded=files.upload()

```

```
import io
import pandas as pd

data = pd.read_csv(io.BytesIO(uploaded['kerala covid main  data.csv']))
```

```
data=data.drop(["Unnamed: 6"],axis=1)
```

```
data=data.drop([14],axis=0)
```

```
data=data.drop([15],axis=0)
data.tail()
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = data['District']

Y = data['Confirmed']
Z = data['Active']
R = data['Rcvrd']


x = np.arange(len(labels))  
width = 0.30 

fig,ax = plt.subplots(figsize=(50,20))


C = ax.bar(x + 0, Y, width,color='b',label='Confirmed')
A = ax.bar(x + 0.25, Z, width,color='r',label='Active')
R = ax.bar(x + 0.50, R , width,color='g',label='Recovered')


ax.set_ylabel('Count',size=45)
ax.set_title('Kerala District wise breakdown',size=45)
ax.set_xticks(x)
ax.set_xticklabels(labels,rotation=90,size=45)
ax.tick_params(axis = 'y', labelsize = 45)
ax.legend(fontsize=45)

plt.show()
```
